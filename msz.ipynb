### Using PyCaret Lib 

```python
# Classification Problem so selecting 

from pycaret.classification import *
clf=setup(data=df, target='IsGoodNews')
```

```python
# Comparing models . It will give the comparative analysis of all the models as below..
compare_models()
```

```python
# Based on the best model choosing LGBM  for model creation in PyCater
lgbm=create_model('lightgbm')
```

```python
# tuning LightGBM Model
tuned_lightgbm = tune_model('lightgbm')
```

```python
# plotting a model
plot_model(lgbm)
```

```python
# Interpreting model
interpret_model(lgbm)
```

```python
# predict on hold-out
pred_holdout = predict_model(lgbm)
```

```python
# finalize a model
lr_final = finalize_model(lgbm)
```

```python
# Reading Test Data
test=pd.read_csv('test.csv')
```

```python
# generate predictions on test data
predictions = predict_model(lr_final, data = test)
```

```python
# The prediction is shown in the last 2nd last Label columns bassed on prob Score
predictions.iloc[:,-2:].head()
```

```python
df_Submission=pd.DataFrame()
df_Submission['IsGoodNews']=predictions['Label']
```
